```python
import pandas as pd
```

```python
import GetOldTweets3 as got
tweetCriteria = got.manager.TweetCriteria().setQuerySearch('#windows')\
                                           .setSince("2018-01-01")\
                                           .setUntil("2019-01-01")
tweet = got.manager.TweetManager.getTweets(tweetCriteria)
```

```python
tweety=[]
for t in tweet:
    tweety.append([t.id,t.text, t.date , t.retweets , t.favorites , t.hashtags])
```

```python
df = pd.DataFrame(tweety, columns =['id','text', 'date','RT','heart','hashtag'])
```

```python
df.to_csv('windows.csv')
```

```python
msft = pd.read_csv('msft.csv')
wind = pd.read_csv('windows.csv')
smsft = pd.read_csv('$msft.csv')
ms = pd.read_csv('microsoft.csv')
dat = pd.concat([msft,wind,smsft,ms])
dat.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
import re

def processTweet(tweet):

    tweet = tweet.lower()
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',tweet)
    tweet = re.sub('@[^\s]+','AT_USER',tweet)
    tweet = re.sub('[\s]+', ' ', tweet)
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    tweet = tweet.strip('\'"')
    return tweet

dat['text'] = dat['text'].apply(lambda line : processTweet(line))
```

```python
dat = dat.drop_duplicates()
```
